```python
import Config
import matplotlib.pyplot as plt
import numpy as np
import torch
import torchvision
from torchvision.utils import save_image
from torchvision.io import read_image
import os
import Test
import dataloader
from torch.utils.data import DataLoader
import albumentations as A
from albumentations.pytorch.transforms import ToTensorV2
from imutils import paths

X_test = []
y_test = []

with open("C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\output\\validation_paths0.txt", "r") as f:
    for path in f:
        filename = os.path.basename(path)
        X_test.append(path.rstrip('\n'))
        y = 'C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\Masks\\'+filename
        y_test.append(y.rstrip('\n'))
f.close()

X_test = np.array(sorted((X_test)))
y_test = np.array(sorted((y_test)))

test_transform = A.Compose([A.Resize(Config.Input_Height, Config.Input_Width),
                            A.Normalize(mean=(0.0), std=(1.0)),
                            ToTensorV2()])

testDS = dataloader.dataset(imagePaths=X_test, maskPaths=y_test, transform=test_transform)
testLoader = DataLoader(testDS, shuffle=False, batch_size=1,
                             pin_memory=Config.PIN_MEMORY, num_workers=os.cpu_count())

model_path = 'C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\output\\unet_out.pth'
model = torch.load(model_path, map_location='cpu')

Test.save_predictions_as_imgs(imagesPath = X_test, loader = testLoader, model = model, 
                              folder = 'C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\output\\Fold 1', device = Config.DEVICE)
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
import cv2
import os
import numpy as np
import pandas as pd

def iou(pred, gt):
    intersection = np.sum((pred == 1) & (gt == 1))
    union = np.sum((pred == 1) | (gt == 1))
    if union == 0:
        return 0
    return intersection / union


# ground truth masks and predictions are binary numpy arrays of shape (num_samples, height, width)
gt_path = "C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\Masks_700"
pred_path = "C:\\Users\\z004b1tz\\Desktop\\Master Thesis Project\\5 Layer Binary UNET\\Dataset\\output\\Fold 1"

scores = {'filename': [], 'Dice score': [], 'IoU': [], 'precision': [], 'Recall': []}

for file in os.listdir(pred_path):
    if file.endswith('.PNG'):
        filename = os.fsdecode(file)
        y_pred = cv2.imread(pred_path+'\\'+filename,cv2.IMREAD_GRAYSCALE)
        y_pred = y_pred.astype(float) / 255
        y_true = cv2.imread(gt_path+'\\'+filename,cv2.IMREAD_GRAYSCALE)
        _, y_true = cv2.threshold(y_true, 128, 255, cv2.THRESH_BINARY)
        y_true = y_true.astype(float) / 255
        # Flatten the arrays to shape (num_samples, height * width)
        y_true_flat = y_true.reshape(-1)
        y_pred_flat = y_pred.reshape(-1)

        precision = precision_score(y_true_flat, y_pred_flat)
        recall = recall_score(y_true_flat, y_pred_flat)
        IoU = iou(y_pred_flat, y_true_flat)
        dice_score = 2 * (y_true_flat * y_pred_flat).sum() / (y_true_flat.sum() + y_pred_flat.sum())
        
        scores['filename'].append(filename)
        scores['Dice score'].append(dice_score)
        scores['IoU'].append(IoU)
        scores['precision'].append(precision)
        scores['Recall'].append(recall)

df = pd.DataFrame(scores)
df
average_f1_score = df['IoU'].mean()
average_precision = df['precision'].mean()
average_recall = df['Recall'].mean()
average_dice_score = df['Dice score'].mean()

print("Average Dice score:", average_dice_score)
print("Average IoU:", average_f1_score)
print("Average precision:", average_precision)
print("Average recall:", average_recall)
```

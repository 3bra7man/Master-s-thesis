```python
boxes = {'filename': '01893_0035_976x976_16bpp.PNG',
  'boxes': [[163.71281, 459.43292, 402.95047, 732.5692],
   [357.68182, 313.0881, 459.932, 807.6449]]}

gt = [[358, 801, 459, 313],
     [162, 705, 406, 451]]

filename = '01893_0035_976x976_16bpp.PNG'

final = {'filename':'', 'mse_error':[], 'manhattan':[], 'mean_abs_error':[]}

final['filename'] = filename
for point in gt:
    mse_errors = []
    euclidean_dists = []
    manhattan_dists = []
    mean_abs_errors = []
    x1_ground, y1_ground = min(point[0], point[2]), min(point[1], point[3])
    x2_ground, y2_ground = max(point[0], point[2]), max(point[1], point[3])
    for box in boxes['boxes']:
        mse_error = np.sqrt((x2_ground - box[2])**2 + (y2_ground - box[3])**2)
        manhattan_dist = np.sum(np.abs(np.array((x2_ground, y2_ground)) - np.array((box[2], box[3]))))
        mean_abs_error = (abs(x2_ground - box[2]) + abs(y2_ground - box[3])) / 2
        mse_errors.append(mse_error)
        euclidean_dists.append(euclidean_dist)
        manhattan_dists.append(manhattan_dist)
        mean_abs_errors.append(mean_abs_error)
    final['mse_error'].append(min(mse_errors))
    final['manhattan'].append(min(manhattan_dists))
    final['mean_abs_error'].append(min(mean_abs_errors))
final['mse_error'] = np.mean(final['mse_error'])
final['manhattan'] = np.mean(final['manhattan'])
final['mean_abs_error'] = np.mean(final['mean_abs_error'])
```
